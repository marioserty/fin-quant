```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.rcParams.update({'font.size': 16}) # Aumenta fonte e números dos eixos nos gráficos
```

# Gaussian White Noise

```python
np.random.seed(123)
y = np.random.normal(0, 1, 250)

plt.figure(figsize=(10, 6))
plt.plot(y, color='blue')
plt.title('Processo de ruido branco Gaussiano'), plt.xlabel('Tempo'), plt.ylabel('y(t)')
plt.axhline(0, color='black') # Adiciona reta horizontal
plt.show()
```

# log returns Gaussian white noise

```python
np.random.seed(123)
y = np.random.normal(0, 0.05, 60)
t = np.arange(1, len(y) + 1) # Vetor X necessário para plots do tipo scatter

plt.figure(figsize=(10, 6))
plt.scatter(x=t, y=y, color='blue', facecolor='none', linewidth=2)
plt.title('Processo GWN para log-retornos mensais'), plt.xlabel('Tempo'), plt.ylabel('y(t)')
plt.axhline(0, color='black')
plt.axhline(0.05, color='red', linestyle='--'), plt.axhline(-0.05, color='red', linestyle='--')
plt.show()
```

# Tendencia deterministica

```python
np.random.seed(123)
e = np.random.normal(0, 1, 250)
y_dt = 0.1 * np.arange(1, 251) + e
abline_values = [0.1 * i for i in np.arange(1, 251)] # Valores da reta crescente (aX + b)

plt.figure(figsize=(10, 6))
plt.plot(abline_values, color='black', linewidth=2)
plt.plot(y_dt, color='blue',linewidth=2)
plt.title('Tendencia deterministica + ruido'), plt.xlabel('Tempo'), plt.ylabel('y dt')
plt.show()
```

# Random Walk

```python
np.random.seed(321)
e = np.random.normal(0, 1, 250)
y_rw = np.cumsum(e)

plt.figure(figsize=(10, 6))
plt.plot(y_rw, color='blue')
plt.title('Passeio aleatorio'), plt.xlabel('Tempo'), plt.ylabel('y.rw')
plt.axhline(0, color='black')
plt.show()
```

# Processo MA(1)

```python
from statsmodels.tsa.arima_process import ArmaProcess
from statsmodels.tsa.stattools import acf
from statsmodels.graphics.tsaplots import plot_acf
```

```python
np.random.seed(12345)
mu = 1
ma1_sim = mu + ArmaProcess(ma=[1, 0.9]).generate_sample(nsample=250) + np.random.normal(0, 0.1, 250)

# FAC
ma1_fac = acf(ma1_sim)

fig, axes = plt.subplots(2, 1, figsize=(12, 8))
axes[0].plot(ma1_sim, linewidth=2)
axes[0].axhline(0, color='black'), axes[0].axhline(1, color='black')
axes[0].set_title('Processo MA(1): mu=1, theta=0.9')
axes[0].set_xlabel('time'), axes[0].set_ylabel('y(t)')

plot_acf(ma1_fac, lags=10, title='FAC para MA(1): theta=0.9', fft=False, ax=axes[1])
axes[1].set_ylabel('FAC'), axes[0].set_xlabel('lag')
plt.tight_layout()
plt.show()
```

# AR(1) simulação

```python
np.random.seed(123)
mu = 1
ar1_sim = mu + ArmaProcess(ar=[1, -0.9]).generate_sample(nsample=250)

# FAC
ar1_fac = acf(ar1_sim)

fig, axes = plt.subplots(2, 1, figsize=(12, 8))
axes[0].plot(ar1_sim, linewidth=2)
axes[0].axhline(0, color='black'), plt.axhline(1, color='black')
axes[0].set_title('Processo AR(1): mu=1, phi=0.9'), axes[0].set_xlabel('time'), axes[0].set_ylabel('y(t)')

plot_acf(ar1_fac, lags=20, title='FAC para AR(1): phi=0.9', fft=False, ax=axes[1])
axes[1].set_ylabel('FAC'), axes[1].set_xlabel('lag')
plt.tight_layout()
```

# Arma(1, 1)

```python
np.random.seed(123)
mu = 1
arma11_sim = mu + ArmaProcess(ar=[1, -0.9], ma=[1, 0.9]).generate_sample(nsample=250)

# FAC
arma11_fac = acf(arma11_sim)

fig, axes = plt.subplots(2, 1, figsize=(12, 8))
axes[0].plot(arma11_sim, linewidth=2)
axes[0].axhline(0, color='black'), axes[0].axhline(1, color='black')
axes[0].set_title('Processo ARMA(1, 1)')
axes[0].set_xlabel('time'), axes[0].set_ylabel('y(t)')

plot_acf(arma11_fac, lags=20, title='FAC para ARMA(1, 1)', fft=False, ax=axes[1])
axes[1].set_ylabel('FAC'), axes[0].set_xlabel('lag')
plt.tight_layout()
plt.show()
```

# Passeio Aleatório e Passeio Aleatório com Tendência

```python
# Passeio aleatorio com tendencia
# Random walk
np.random.seed(321)
e = np.random.normal(0, 1, 250)
y_rw = np.cumsum(e)

plt.figure(figsize=(12, 8))
plt.subplot(211)
plt.plot(y_rw, color='blue')
plt.title('Passeio aleatorio'), plt.xlabel('Tempo'), plt.ylabel('y.rw')
plt.axhline(0, color='black')

# Random walk with drift
plt.subplot(212)
mu = np.repeat(0.1, len(e))
y_rwd = np.cumsum(mu) + np.cumsum(e)

plt.plot(y_rwd, color='blue')
plt.title('Passeio aleatorio com tendência'), plt.xlabel('Tempo'), plt.ylabel('y.rwd')
plt.axhline(0, color='black')

plt.tight_layout()
plt.show()
```

# Teste de I(0)

```python
import yfinance as yf # Baixar dados do yahoo fianance
data = yf.download('^BVSP', start='2000-01-01', end='2020-08-01')
data.head()
```

```python
data = data['Close']
data.fillna(method='ffill', inplace=True) # Substitui valores NA por valor anterior
returns = data.pct_change().dropna()
returns.head()
```

```python
from statsmodels.tsa.stattools import adfuller

nivel_significancia = 0.05
teste_adf = adfuller(returns)
p_valor = teste_adf[1]

if (p_valor < nivel_significancia):
    print('A série é estacionária')
else:
    print('A série não é estacionária')
```

```python
plt.figure(figsize=(12, 12))

plt.subplot(411)
returns.plot(title='IBOV retornos entre 2000 e 2020')

i1 = returns.cumsum()
plt.subplot(412)
i1.plot(title='IBOV: série I(1), 2000 e 2020')

i2 = i1.cumsum()
plt.subplot(413)
i2.plot(title='IBOV: série I(2), 2000 e 2020')

i3 = i2.cumsum()
plt.subplot(414)
i3.plot(title='IBOV: série I(3), 2000 e 2020')

plt.tight_layout()
```

# Cointegração

```python
petr4 = yf.download('PETR4.SA', start='2013-01-01', end='2017-08-01')
petr4 = petr4['Close'].fillna(method='ffill')
petr3 = yf.download('PETR3.SA', start='2013-01-01', end='2017-08-01')
petr3 = petr3['Close'].fillna(method='ffill')

petr4_i0 = petr4.pct_change().dropna()
petr3_i0 = petr4.pct_change().dropna()
diferenca = petr3 - petr4

series = [petr4, petr4_i0, petr3, petr3_i0, diferenca]
names = ['PETR4', 'PETR4 I(0)', 'PETR3', 'PETR3 I(0)', 'Diferença']

for s, n in zip(series, names):
    print('---')
    
    teste_adf = adfuller(s, autolag='BIC')
    p_valor = teste_adf[1]

    if (p_valor < nivel_significancia):
        print(f'A série "{n}" é estacionária, p-valor = {p_valor}')
    else:
        print(f'A série "{n}" NÃO é estacionária, p-valor = {p_valor}')
```

```python
plt.figure(figsize=(12, 8))
petr4.plot(color='r', label='PETR$')
petr3.plot(color='b', label='PETR3')
diferenca.plot(color='g', label='Diferença')
plt.axhline(0, color='black')
plt.ylabel('Preços'), plt.title('PETR3, PETR4 e diferença entre 2013 e 2017')
plt.legend()
```
