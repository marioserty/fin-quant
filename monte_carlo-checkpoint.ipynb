```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
```

# Data Reading

```python
# Table downloaded from http://www.dcc.ufmg.br/~arbex/portfolios/IBOV.csv
prices = pd.read_csv('IBOV.csv', parse_dates=['Date'], index_col='Date')
prices[prices.columns[:10]].tail()
```

# Get IBOV Prices

```python
ibov = prices.loc['2017':'2018-01-02']['IBOV']
ibov.tail()
```

# Get IBOV returns

```python
returns = ibov.pct_change().dropna()
returns.tail()
```

```python
plt.figure(figsize=(12, 4))
plt.plot(returns)
plt.title('IBOV daily returns'), plt.xlabel('Date'), plt.ylabel('Return (%)')
plt.show()

plt.figure(figsize=(12, 4))
plt.plot(returns.add(1).cumprod().sub(1))
plt.title('IBOV cumulative returns'), plt.xlabel('Date'), plt.ylabel('Return (%)')
plt.show()
```

# Simulation

```python
# First historic VaR assuming normal distribution
rmu, rsd = returns.mean(), returns.std()
var_5 = norm.ppf(q=0.05, loc=rmu, scale=rsd)
print('Mean = {:.5f}, SD = {:.5f}, VaR 5% = {:.5f}'.format(rmu, rsd, var_5))
```

```python
# Index price if the return is equal to VaR:
historic_var_price = ibov[-1]*(1+var_5)
historic_var_price
```

```python
# Monte Carlo, using Constant Expected Return model with epsilon ~ N(0, 1)

# simulations = []
# for i in range(10000):
#     simulation_return = rmu + rsd*np.random.normal(0, 1)
#     simulations.append(ibov[-1] * (1 + simulation_return))

simulations = ibov[-1] * (1 + (rmu + rsd * np.random.normal(0, 1, 10000))) # Simplified

simulations = sorted(simulations)
var_idx  = int(len(simulations) * 0.05) + 1
simulated_var_price = simulations[var_idx]

print("Last value         :", ibov[-1])
print("Historic VaR Value :", historic_var_price)
print("Simulated VaR Value:", simulated_var_price)
```
